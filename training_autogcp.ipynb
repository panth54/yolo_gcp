```python
from ultralytics import YOLO

# Load a YOLOv8 model
model = YOLO('yolov8n.pt')  # 'n' for nano, use smaller models for small datasets

# Train the model with 456x456 images
results = model.train(
    data='dataset.yaml', 
    epochs=50,  # Use fewer epochs to prevent overfitting
    imgsz=456, 
    batch=8,  # Smaller batch size for small datasets
    device=0,  # Use GPU (0) or CPU (-1)
    augment=True  # Enable data augmentation
)
```

```python
# Evaluate the model
val_results = model.val()
```

```python
# Save the trained model
model.save('trained_model.pt')
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load the trained model
model = YOLO('trained_model.pt')  

# Make predictions on a sample image
img_path = 'image_3.jpg'  # Replace with the path to your image
results = model(img_path)

# Load the image using OpenCV
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert color from BGR to RGB

# Extract bounding boxes and labels from the results
for result in results:
    boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding boxes
    scores = result.boxes.conf.cpu().numpy()  # Get confidence scores
    classes = result.boxes.cls.cpu().numpy()  # Get class labels

    # Draw bounding boxes on the image
    for box, score, cls in zip(boxes, scores, classes):
        x1, y1, x2, y2 = box
        label = f'Class {int(cls)}: {score:.2f}'
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Display the image with Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load the trained model
model = YOLO('trained_model.pt')  

# List of image paths
image_paths = ['image_3.jpg', 'image_5.jpg']  # Replace with the paths to your images

# Create a figure for displaying images
fig, axs = plt.subplots(1, len(image_paths), figsize=(15, 10))

for ax, img_path in zip(axs, image_paths):
    # Make predictions on the image
    results = model(img_path)

    # Load the image using OpenCV
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert color from BGR to RGB

    # Extract bounding boxes and labels from the results
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding boxes
        scores = result.boxes.conf.cpu().numpy()  # Get confidence scores
        classes = result.boxes.cls.cpu().numpy()  # Get class labels

        # Draw bounding boxes on the image
        for box, score, cls in zip(boxes, scores, classes):
            x1, y1, x2, y2 = box
            label = f'Class {int(cls)}: {score:.2f}'
            cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the image with Matplotlib
    ax.imshow(image)
    ax.axis('off')  # Hide axis
    ax.set_title(f'Prediction for {img_path}')

plt.tight_layout()
plt.show()
```
